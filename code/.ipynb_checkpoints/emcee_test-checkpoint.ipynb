###Front page example

```python
import numpy as np
import emcee
%matplotlib inline
```

```python
#from http://dan.iel.fm/emcee/current/

def lnprob(x, ivar):
    return -0.5 * np.sum(ivar * x ** 2)

ndim, nwalkers = 10, 100
ivar = 1. / np.random.rand(ndim)
p0 = [np.random.rand(ndim) for i in range(nwalkers)]

sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob, args=[ivar])
sampler.run_mcmc(p0, 10)
print 'done'
```

###Another example - quickstart  
Multidimensional Gaussian:  
$\rho(x) \propto \exp\left(-0.5(x-\mu)\Sigma^{-1}(x-\mu)\right)$  
i.e.  
$\rho(x) = \frac{1}{\sqrt{(2\pi)^k|\Sigma|}} \exp\left(-0.5(x-\mu)\Sigma^{-1}(x-\mu)\right)$  
where $k$ is the dimension.

```python
#quickstart one
#note - emcee requires the log of the density!
def lnprob(x, mu, inv_cov):
    difference = x-mu
    return -np.dot(difference,np.dot(inv_cov,difference))/2.0
```

```python
ndim = 50
means = np.random.rand(ndim)

cov = 0.5 - np.random.rand(ndim ** 2).reshape((ndim, ndim)) #what's going on here?
cov = np.triu(cov)
cov += cov.T - np.diag(cov.diagonal())
cov = np.dot(cov,cov)

inv_cov = np.linalg.inv(cov)
```

```python
nwalkers = 250
p0 = np.random.rand(ndim * nwalkers).reshape((nwalkers, ndim))
```

```python
#number of walkers, dimension of problem/parameter set, probability function, extra parameters
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob, args=[means, inv_cov])
```

```python
#burn in
pos, prob, state = sampler.run_mcmc(p0, 100)
sampler.reset()
```

```python
#actual run, from burn in
sampler.run_mcmc(pos, 1000)
print 'done'
```

```python
import matplotlib.pyplot as plt

n_to_plot= int(0.1*ndim)
for i in range(n_to_plot):
    plt.figure()
    plt.hist(sampler.flatchain[:,i], 100, color="k", histtype="step")
    plt.title("Dimension {0:d}".format(i))

plt.show()
```

```python
print("Mean acceptance fraction: {0:.3f}"
                .format(np.mean(sampler.acceptance_fraction)))
```

```python
#save sampler chains
#see pickle website for more!
import pickle

pickle.dump(sampler.flatchain, open( "save_sampler_flat_chain.p", "wb" ) )
```

```python
#get sampler back (test by restarting kernel)
import emcee
import pickle

sampler_flat_chain = pickle.load( open( "save_sampler_flat_chain.p", "rb" ) )
```

```python
import matplotlib.pyplot as plt
ndim=50
n_to_plot= int(0.1*ndim)
for i in range(n_to_plot):
    plt.figure()
    plt.hist(sampler_flat_chain[:,i], 100, color="k", histtype="step")
    plt.title("Dimension {0:d}".format(i))

plt.show()
```

##Using 'triangle plot'  
install via pip install triangle_plot

```python
#demo.py from triangle github
#note - not a plot of the above
%matplotlib inline
import numpy as np
import triangle
# Set up the parameters of the problem.
ndim, nsamples = 3, 50000
# Generate some fake data.
data1 = np.random.randn(ndim * 4 * nsamples / 5).reshape([4 * nsamples / 5,
ndim])
data2 = (5 * np.random.rand(ndim)[None, :]
+ np.random.randn(ndim * nsamples / 5).reshape([nsamples / 5, ndim]))
data = np.vstack([data1, data2])
# Plot it.
figure = triangle.corner(data, labels=[r"$x$", r"$y$", r"$\log \alpha$",
r"$\Gamma \, [\mathrm{parsec}]$"],
truths=[0.0, 0.0, 0.0],
quantiles=[0.16, 0.5, 0.84],
show_titles=True, title_args={"fontsize": 12})
figure.gca().annotate("A Title", xy=(0.5, 1.0), xycoords="figure fraction",
xytext=(0, -5), textcoords="offset points",
ha="center", va="top")
figure.savefig("demo.png")
```

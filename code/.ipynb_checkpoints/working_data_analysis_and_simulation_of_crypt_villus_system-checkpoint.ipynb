```python
def velocity_hollings_iv(p=np.array([1.0,1.0,-1.0,1.0])):
    """
    simple increasing hill velocity function
    ax^d/(b+cx+x^d) 
    """
    a,b,c,d= p #note neat assignment!
    #velocity function
    if (a >=0) and (b>=0) and (d>=0.0): #(c <= 0.0)
        vel= lambda x: a*(x**d)/(b+c*x+x**d)#vel. is time-independent for whole problem.
    else:
        vel= lambda x: 0.0*x
    return vel
```
